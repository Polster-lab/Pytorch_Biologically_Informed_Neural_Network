```python
clinical_data = pd.read_csv('../../common_data/ROSMAP_DATA/RSP_CL.csv')
```

```python
merged_metadata = pd.merge(individual_meta_data, clinical_data, on='individualID', how='inner', suffixes=('_df1', '_df2'))
```

```python
merged_metadata = merged_metadata[['individualID', 'individualIdSource', 'sex',
       'ethnicity',
        'subject', 'projid', 'Study',
       'msex', 'educ', 'spanish', 'apoe_genotype',
       'age_at_visit_max', 'age_first_ad_dx', 'age_death',
       'cts_mmse30_first_ad_dx', 'cts_mmse30_lv', 'braaksc',
       'ceradsc', 'cogdx', 'dcfdx_lv']]
```

```python
merged_metadata
```

```python
def return_clinical_AD(x):
    if x == 1.0:
        return 'NCI'
    elif x == 4.0:
        return 'AD'
    elif x == 5.0:
        return 'AD'
    else:
        return False
```

```python
merged_metadata['clinical_diagnosis'] = merged_metadata.cogdx.apply(return_clinical_AD)
```

```python
merged_metadata
```

```python
def return_pathological_AD(x):
    if x == 1.0:
        return 'AD'
    elif x == 4.0:
        return 'No AD'
    else:
        return False
```

```python
merged_metadata['pathological_diagnosis'] = merged_metadata.ceradsc.apply(return_pathological_AD)
```

```python
merged_metadata
```

```python
def return_clinical_pathological_AD(x):
    if x['cogdx'] == 1.0 and x['ceradsc'] == 4.0:
        return 'NCI_with_No_Plaques'
    elif x['cogdx'] == 4.0 and x['ceradsc'] == 1.0:
        return 'AD_with_Plaques'
    else:
        return False
```

```python
merged_metadata['clinical_pathological_AD'] = merged_metadata.apply(return_clinical_pathological_AD,axis=1)
```

```python
import numpy as np
def return_age_death(x):
    if x == '90+':
        return 90
    else:
        return np.float64(x)

merged_metadata['age_death'] = merged_metadata.age_death.apply(return_age_death)
```

```python
merged_metadata.to_csv('/12tb_dsk1/danish/preprocessed_data/clinical/clinical_single_cell.csv', index = False)
```
