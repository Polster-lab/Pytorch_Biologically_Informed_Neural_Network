```r
individual_metadata <- read.delim("../../../common_data/single_cell_data/individual_metadata_deidentified.tsv", stringsAsFactors = FALSE)
head(individual_metadata,5)
```

```r
# Load necessary libraries

library(dplyr)

# Load your RDS object


# Read the TSV file

# Ensure the subject columns are character type (if they are not already)
rds_obj@meta.data$subject <- as.character(rds_obj@meta.data$subject)
individual_metadata$subject <- as.character(individual_metadata$subject)

# Merge the diagnosis information into the meta.data slot of the RDS object
rds_obj@meta.data <- rds_obj@meta.data %>%
  left_join(individual_metadata %>% select(subject, Pathologic_diagnosis_of_AD), by = "subject")

# Check if the merge was successful
head(rds_obj@meta.data)
```

```r
microglia_Mic_P2RY12_data <- rds_obj@meta.data %>% filter(cell_type_high_resolution == 'Mic P2RY12')
head(microglia_Mic_P2RY12_data,10)
```

```r
table(microglia_Mic_P2RY12_data$Pathologic_diagnosis_of_AD)
```

```r
print(head(count_matrix, 5))
```

```r
cell_ids_microglia_Mic_P2RY12 <- microglia_Mic_P2RY12_data$cell_id
head(cell_ids_microglia_Mic_P2RY12,10)
```

```r
logical_vector <- colnames(count_matrix) %in% cell_ids_microglia_Mic_P2RY12
head(logical_vector,10)
```

```r
selected_count_matrix <- count_matrix[, logical_vector]
dim(selected_count_matrix)
```

```r
print(head(selected_count_matrix, 5))
```

```r
microglia_seurat_obj <- CreateSeuratObject(counts = selected_count_matrix, project = "ROSMAP", min.cells = 0, min.features = 0)
```
